### **`7. Introduction to Seaborn`**

In this study material, we will explore the installation process of Seaborn, discuss its role in statistical data visualization, and provide an overview of its capabilities.

#### c. Advanced Heatmaps for Correlation

Heatmaps are an effective way to visualize the correlation matrix of variables. Seaborn's `sns.heatmap` provides customization options and an easy-to-use interface for creating insightful heatmaps.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(42)  # For reproducibility
data = pd.DataFrame({
    'Feature1': np.random.rand(100),
    'Feature2': np.random.rand(100),
    'Feature3': np.random.rand(100),
    'Target': np.random.rand(100)
})

# Correlation heatmap with Seaborn
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
````

***Explantaion:***

`np.random.seed()` is a function from the NumPy library in Python that is used to set the seed for the random number generator. When you set the seed using `np.random.seed()`, it ensures that if you run the same code multiple times, you will get the same random numbers each time. This is useful for reproducibility, especially in scenarios like generating random data or splitting datasets into training and testing sets, where you want the randomness to be consistent across runs.

In the provided example:

```python
np.random.seed(42)
```

The seed is set to the value `42`. You can choose any integer value for the seed. The key is that if you set the seed to the same value, you'll get the same sequence of random numbers, allowing you to reproduce the exact results.



The above code generates a correlation heatmap using Seaborn to visualize the correlation between variables in a dataset. Let's break down the code:

1. `correlation_matrix = data.corr()`: Calculates the correlation coefficients between all pairs of numerical columns in the DataFrame `data`. The result is a correlation matrix where each element (i, j) represents the correlation coefficient between the i-th and j-th columns of the DataFrame.

2. `sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')`:
   - `correlation_matrix`: The correlation matrix calculated in the previous step.
   - `annot=True`: Adds numeric annotations to each cell of the heatmap, displaying the exact correlation values.
   - `cmap='coolwarm'`: Specifies the color map to be used for the heatmap. In this case, the 'coolwarm' color map is chosen, which represents low correlations in cool colors (blue) and high correlations in warm colors (red).

3. `plt.show()`: Displays the generated heatmap using Matplotlib.

The resulting heatmap visually represents the correlation structure of the variables in the dataset. High positive correlations are displayed in warmer colors, while high negative correlations are shown in cooler colors. The annotation values provide precise correlation coefficients for each pair of variables.

This type of visualization is helpful for understanding relationships between variables and identifying patterns in the data. High absolute values of correlation coefficients (close to 1 or -1) indicate strong relationships, while values close to 0 suggest weaker or no linear correlation.

Seaborn's user-friendly syntax and versatile capabilities make it a powerful tool for statistical data visualization, making complex plots accessible even to users with less experience in data visualization.

By combining the strengths of Seaborn with other data analysis libraries, you can create compelling visualizations that help uncover patterns, trends, and relationships within your datasets.

### **`8. Categorical Plots with Seaborn`**

Seaborn provides a variety of categorical plot types that are essential for visualizing relationships and distributions within different categories of data. In this section, we'll delve into the creation of categorical scatter plots, bar plots, box plots, and violin plots using Seaborn. Real-world examples will be provided to illustrate the scenarios where these plots are most useful.

#### 1. Categorical Scatter Plots:

**Syntax:**
```python
sns.stripplot(x='category_column', y='numeric_column', data=data)
```

**Usage:**
Categorical scatter plots are particularly useful when you want to visualize the distribution of numeric data across different categories. Each point in the plot represents an individual data point, providing a clear overview of the data's spread within categories.

**Real-world Example:**
Consider a dataset with student scores (`numeric_column`) across different subjects (`category_column`). A categorical scatter plot can be employed to visualize individual student scores in each subject.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generating example student data
np.random.seed(42)  # Setting seed for reproducibility

subjects = ['Math', 'Science', 'History', 'English']
student_data = pd.DataFrame({
    'subject': np.random.choice(subjects, size=100),
    'score': np.random.randint(60, 100, size=100)  # Random scores between 60 and 100
})

# Creating a categorical scatter plot
sns.stripplot(x='subject', y='score', data=student_data)
plt.title('Student Scores Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
````

**More about `sns.stripplot()`**

The `sns.stripplot()` function in Seaborn is used to create a categorical scatter plot. It is particularly useful when you want to visualize individual data points along a single axis for different categories.

Here's the breakdown of the function:

```python
sns.stripplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=True, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None)
```

- `x`, `y`: The variables specifying the data on the x and y-axis.
- `hue`: Grouping variable that will produce points with different colors. It can be a categorical variable.
- `data`: DataFrame or array-like object.
- `order`, `hue_order`: Order to plot the categorical levels in, otherwise, the levels are inferred from the data objects.
- `jitter`: Adds random noise to the categorical scatter plot points to avoid overlapping.
- `dodge`: When hue nesting is used, setting this to `True` will separate the strips for different hue levels along the categorical axis.
- `orient`: Orientation of the plot (`'v'` for vertical, `'h'` for horizontal).
- `color`: Color for all of the elements, or seed for a gradient palette.
- `palette`: Colors to use for the different levels of the hue variable.
- `size`: Diameter of the markers, in points.
- `edgecolor`: Color of the lines around each point.
- `linewidth`: Width of the lines around each point.
- `ax`: Matplotlib Axes, optional.



````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = sns.load_dataset("tips")

# Creating a stripplot
sns.stripplot(x="day", y="total_bill", data=data, jitter=True, hue="sex", dodge=True)

# Adding a legend
plt.legend(title="Sex", loc="upper left")

plt.show()
````


In this example, the `stripplot` is used to visualize the distribution of total bills (`total_bill`) across different days (`day`). The `jitter=True` adds random noise to avoid overlap, and `hue="sex"` differentiates the points by gender. The `dodge=True` separates the strips for different gender levels along the categorical axis.

#### 2. Bar Plots:

**Syntax:**
```python
sns.barplot(x='category_column', y='numeric_column', data=data)
```

**Usage:**
Bar plots are effective for summarizing the central tendency of numeric data for different categories. They display the mean or another estimator and the confidence interval around it, providing a quick visual comparison of values.

**Real-world Example:**
Imagine you have restaurant data with total bill amounts (`numeric_column`) and the days on which the bills were recorded (`category_column`). A bar plot can help illustrate the average total bill for each day.product.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data using the "tips" dataset from seaborn
restaurant_data = sns.load_dataset("tips")

# Creating a bar plot
sns.barplot(x='day', y='total_bill', data=restaurant_data)
plt.title('Average Total Bill Across Different Days')
plt.show()
````

In this example, the "day" column represents different days of the week, and the "total_bill" column signifies the total bill amounts recorded on those days. The resulting bar plot provides insights into the average total bill amounts for each day of the week based on the restaurant data.

**More about `sns.barplot()`**

The `sns.barplot()` function in Seaborn is used to create a bar plot. It is a categorical plot that displays the estimate of a central tendency (usually the mean) for a numeric variable within different levels of one or more categorical variables. This plot provides a good way to visualize the relationship between a categorical variable and a numerical variable.

Here is the general syntax:

```python
sns.barplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, ci='sd', estimator=<function mean at 0x...>, n_boot=1000, units=None, seed=None, orient=None, color=None, palette=None, saturation=0.75, errcolor='.26', errwidth=None, capsize=None, dodge=True, ax=None, **kwargs)
```

- `x`, `y`: The names of variables in your dataset for the x and y axes.
- `hue`: A variable in your data to map to the color of the bars. It adds a nested categorical variable to the plot.
- `data`: The DataFrame or data structure to use.
- `order`, `hue_order`: Order to plot the categorical levels in, otherwise, it is inferred from the data.
- `ci`: Confidence interval for the mean. It can take values such as `'sd'` for standard deviation, or `None` to disable confidence intervals.
- `estimator`: Function to estimate the central tendency (default is mean).
- `n_boot`: Number of bootstrap iterations to compute confidence intervals.
- `units`: Identifier of sampling units if data should be treated as if it were a multilevel dataset.
- `orient`: Orientation of the plot (`'v'` for vertical, `'h'` for horizontal).
- `color`, `palette`: Color for the bars.
- `saturation`: Proportion to desaturate the plot.
- `errcolor`, `errwidth`: Color and width of the error bars.
- `capsize`: Width of the caps on error bars.
- `dodge`: When hue nesting is used, whether elements should be shifted along the categorical axis.
- `ax`: Axes in which to draw the plot.
- `**kwargs`: Additional keyword arguments passed to the function.

Here's a simple example using the `tips` dataset:

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
tips_data = sns.load_dataset("tips")

# Creating a bar plot
sns.barplot(x='day', y='total_bill', data=tips_data)
plt.title('Average Total Bill Across Different Days')
plt.show()
````

In this example, the bar plot shows the average total bill for different days (`x='day'`, `y='total_bill'`). You can customize it further based on your specific requirements and dataset.

#### 3. Box Plots:

**Syntax:**
```python
sns.boxplot(x='category_column', y='numeric_column', data=data)
```

**Usage:**
Box plots are valuable for visualizing the distribution of numeric data across different categories. They provide a summary of the central tendency, quartiles, and potential outliers.

**Real-world Example:**
Imagine you have a dataset that records the cut quality (`category_column`) of different diamonds and their prices (`numeric_column`). To discern variations in diamond prices across different cut qualities, a box plot can be a valuable visualization tool.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data using the "diamonds" dataset from seaborn
diamond_data = sns.load_dataset("diamonds")

# Creating a box plot
sns.boxplot(x='cut', y='price', data=diamond_data)
plt.title('Diamond Prices Across Different Cut Qualities')
plt.show()
````


In this example, the "cut" column represents different diamond cut qualities, and the "price" column signifies the diamond prices. The resulting box plot provides insights into the distribution and variations in prices based on different cut qualities.

**More about `sns.boxplot()`**

The `sns.boxplot()` function in seaborn is used to create a box plot, also known as a whisker plot. Box plots are useful for visualizing the distribution of a dataset and identifying summary statistics such as the median, quartiles, and potential outliers.

#### Syntax:

```python
sns.boxplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, width=0.8, dodge=True, fliersize=5, linewidth=None, whis=1.5, ax=None, **kwargs)
```

#### Parameters:

- `x`, `y`: Input data variables (either categorical or numerical).
- `hue`: Grouping variable that will produce boxes with different colors.
- `data`: DataFrame or array-like.
- `order`, `hue_order`: Order to plot the categorical levels in, otherwise, they are inferred from the data.
- `orient`: Orientation of the plot (`'v'` for vertical, `'h'` for horizontal).
- `color`: Color for all the elements, or seed for a gradient palette.
- `palette`: Colors to use for the different levels of the hue variable.
- `width`: Width of the boxes.
- `dodge`: When hue nesting is used, whether elements should be shifted along the categorical axis.

#### Usage:

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data (using the "tips" dataset from seaborn)
data = sns.load_dataset("tips")

# Creating a box plot
sns.boxplot(x='day', y='total_bill', data=data, hue='time')
plt.title('Box Plot of Total Bill Across Different Days with Time Variation')
plt.show()
````

***Explanation:***


- `x`, `y`: In this example, 'day' is specified as the categorical variable for the x-axis, and 'total_bill' is the numerical variable for the y-axis.
- `hue`: The 'time' parameter introduces a third dimension, representing a categorical variable that colors the boxes differently. In this case, it differentiates between lunch and dinner times.
- `data`: The "tips" dataset from seaborn, containing information about restaurant tips, is used as the source of the data.
- The box plot visually represents the distribution of total bills across different days, with an additional variation denoted by different colors for lunch and dinner times.

Replace "tips," 'day', 'total_bill', and 'time' with the actual dataset name, categorical column, numerical column, and additional categorical column from your dataset.

The resulting box plot showcases the distribution of total bills, with each box representing the interquartile range (IQR), the line inside the box as the median, and whiskers extending to show the range of the data. The 'hue' parameter enhances the visualization by highlighting variations due to different times of the day, offering a more detailed insight into the dataset.

#### 4. Violin Plots:

**Syntax:**
```python
sns.violinplot(x='category_column', y='numeric_column', data=data)
```

**Usage:**
Violin plots combine aspects of box plots and kernel density plots, providing a comprehensive view of the distribution of numeric data across categories.

**Real-world Example:**
Imagine you have a dataset capturing the dining experience, including the total bill (`numeric_column`) for different days (`category_column`). A violin plot can provide valuable insights into the distribution of total bills across these days.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data using the "tips" dataset from seaborn
event_data = sns.load_dataset("tips")

# Creating a violin plot
sns.violinplot(x='day', y='total_bill', data=event_data)
plt.title('Distribution of Total Bills Across Different Days')
plt.show()
````

In this  example, the data is sourced from the "tips" dataset, which contains information related to restaurant bills. The 'day' column represents different days of the week, while the 'total_bill' column denotes the total amount paid. The violin plot visualizes the distribution of total bills, providing a comprehensive view of the spending patterns across various days.

**More about `sns.violinplot()`**

The `sns.violinplot()` function in Seaborn is used to create a violin plot. A violin plot is useful for visualizing the distribution of a numerical variable across different categories or groups. It combines aspects of a box plot and a kernel density plot, providing a more detailed representation of the data distribution.

Here is an overview of the parameters for the `sns.violinplot()` function:

```python
sns.violinplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, bw='scott', cut=2, scale='area', scale_hue=True, gridsize=100, width=0.8, inner='box', split=False, dodge=True, orient=None, linewidth=None, color=None, palette=None, saturation=0.75, ax=None, **kwargs)
```

- `x`, `y`: Variables that specify the data for the x and y axes.
- `hue`: Variable that produces different colors for the violins.
- `data`: DataFrame or array-like object containing the data.
- `order`, `hue_order`: Order to plot the categorical levels.
- `bw`: Bandwidth of the kernel density estimate (smaller value results in a smoother curve).
- `cut`: Distance, in bandwidth units, to extend the density past the extreme data points.
- `scale`: Method used to scale the width of each violin (`'area'`, `'count'`, `'width'`).
- `scale_hue`: Whether to scale the width based on the number of levels in the hue variable.
- `gridsize`: Number of points in the discrete grid used for computing the kernel density estimate.
- `width`: Width of the violins.
- `inner`: Type of plot to draw inside the violins (`'box'`, `'quartile'`, `'point'`, `'stick'`, `None`).
- `split`: Whether to draw separate violins for each level of the hue variable.
- `dodge`: Whether to dodge the violins.
- `orient`: Orientation of the plot (`'v'` for vertical, `'h'` for horizontal).
- `linewidth`: Width of the gray lines that frame the plot elements.

The violin plot provides a summary of the distribution of the data, showing key features like the median, quartiles, and overall shape. It's particularly useful for comparing the distribution of a numerical variable across different categories in a dataset.

### **`9. Statistical Relationships with Seaborn`**

#### **`Pair Plots in Seaborn for Multivariate Analysis`**

#### Introduction:
Pair plots in Seaborn provide a powerful visualization tool for analyzing the relationships between multiple variables in a dataset. These plots are particularly useful when dealing with several numerical variables, enabling us to explore pairwise interactions and identify patterns.

#### How to Create Pair Plots:
To create a pair plot, Seaborn's `pairplot()` function is employed. This function generates a grid of scatterplots for each pair of variables and histograms for each individual variable along the diagonal.

#### Example Code:

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data (replace this with your actual dataset)
data = sns.load_dataset("iris")

# Creating a pair plot
sns.pairplot(data, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
````

#### Interpretation:
- **Scatterplots:** The scatterplots in the upper and lower triangles represent the relationships between pairs of variables. Each point in a scatterplot corresponds to an observation in the dataset, and the color (in this case, the species) can be used to distinguish different groups.
  
- **Diagonal Plots:** The diagonal plots show histograms or kernel density estimates for individual variables. They provide insights into the univariate distribution of each variable.

#### **`Correlation Heatmaps for Statistical Relationships`**

#### Introduction:
Correlation heatmaps are essential tools for understanding the statistical relationships between numerical variables in a dataset. They visualize the correlation coefficients between variables, highlighting patterns of positive or negative associations.

#### How to Create Correlation Heatmaps:
The `sns.heatmap()` function in Seaborn is used to create correlation heatmaps. This function takes a correlation matrix as input and displays the correlation values as colors, making it easy to identify strong or weak relationships.

#### Example Code:

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data (replace this with your actual dataset)
data = sns.load_dataset("iris")

# Calculating the correlation matrix
correlation_matrix = data.corr()

# Creating a correlation heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
````

#### Interpretation:
- **Color Representation:** Each cell in the heatmap is colored based on the correlation coefficient. Positive correlations are often represented in warm colors (e.g., red), while negative correlations are in cool colors (e.g., blue).
  
- **Annotation:** Annotations within the cells display the correlation values, providing a quantitative understanding of the relationships.

By incorporating pair plots and correlation heatmaps into exploratory data analysis, practitioners gain valuable insights into the underlying statistical relationships within their datasets.
